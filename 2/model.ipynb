```python
%matplotlib inline
from thetis import*
```

```python
lx,ly,nx,ny = 1000.,20.,100.,10.
mesh = RectangleMesh(lx,ly,nx,ny)
```

```python
P1DG = FunctionSpace(mesh,'CG', 1)
bathymetry2d = Function(P1DG, name='bathymetry')
depth = 2.0
bathymetry2d.assign(depth)
```

```python
t_end = 3000.
t_export = 10.
```

```python
solver_obj = solver2d.FlowSolver2d(mesh, bathymetry2d)
options = solver_obj.options
```

```python
options.simulation_end_time = t_end
options.simulation_export_time = t_export
options.timestepper_type ='CrankNicolson'
options.timestep = 10.
options.fields_to_export = ['elev_2d', 'uv_2d']
options.fields_to_export_hdf5 = ['elev_2d', 'uv_2d']
```

```python
init_elev = Function(P1DG, name='init_elevation')
#xy = SpatialCoordinate(mesh)
#gauss_width = 4000.
#gauss_ampl = 2.0
#gauss_expr = gauss_ampl * exp(-((xy[0]-lx/2)/gauss_width)**2)
#init_elev.interpolate(gauss_expr)
xy = mesh.coordinates.dat.data
for i,xy in enumerate (xy):
    if 80 < xy[0] < 100:
        init_elev.dat.data[i] = 5.0
    else:
        init_elev.dat.data[i] = 0.0
```
